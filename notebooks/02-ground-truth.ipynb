# 02 - Ground Truth Generation

Generate Q&A pairs from documentation chunks for retrieval evaluation.

```python
import json, os, sys
sys.path.insert(0, "..")
from notebooks.ground_truth import generate_ground_truth
```

```python
chunks_path = "../data/processed/fastapi/chunks.json"
assert os.path.exists(chunks_path), "Run ingestion first"
with open(chunks_path) as f:
    chunks = json.load(f)
print(f"Loaded {len(chunks)} chunks")
```

```python
ground_truth = generate_ground_truth(chunks[:20])
print(f"Generated {len(ground_truth)} Q&A pairs")
for gt in ground_truth[:3]:
    print(f"  Q: {gt.get('question', '?')[:80]}...  ->  {gt.get('relevant_chunk_id', '?')}")
```

```python
os.makedirs("../data", exist_ok=True)
out_path = "../data/ground_truth.jsonl"
with open(out_path, "w") as f:
    for item in ground_truth:
        f.write(json.dumps(item) + "\n")
print(f"Saved to {out_path}")
```
